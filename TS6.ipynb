#### 1. Demostrar que la función de transferencia $T(S) = \frac{V_2(S)}{V_1(S)}$ se corresponde con una sección bicuadrática de segundo orden (SOS) $T(s)=k⋅\frac{S^2+S⋅\frac{ω_{0Z}}{Q_Z}+ω_{0Z}^2}{S^2+S⋅\frac{ω_{0P}}{QP}+ω_{0P}^2}$.

La función de Transferencia es un número asociado a la admitancia o impedancia entre dos puntos del circuito, y se calcula como el cociente entre las diferencias de potencial entre ambos puntos. Para determinar la función de transferencia que caracteriza a un filtro necesitamos analizar sus componentes en el dominio de la impedancia $(Z)$ o en el de su inversa, la admitancia $(Y)$. De esta manera podemos definir a la función de Transferencia en el dominio complejo como $ T(s) = \frac{Z_{2}}{Z_{1}+Z_{2}} $, con $Z_1$ la suma de las impedancias en serie y $Z_2$ la suma de las impedancias en derivación. Mismamente, ya que $Y = \frac{1}{Z}$ podemos calcular $T(s)$ como $\frac{Y_1}{Y_1 + Y_2}$. El enunciado nos da información sobre las componentes capacitivas, inductivas y resistivas en serie y en paralelo del filtro, como se ve en la siguiente imagen:



Para esta distribución es conveniente calcular $T(s)$ en admitancias. La componente $Y_1$ es $S \cdot C \cdot a + \frac{b}{S\cdot L} + \frac{d}{R}$ y la componente $Y_2$ es $S \cdot C \cdot (1-a) + \frac{1-d}{R} + \frac{1-b}{S \cdot L}$. Componer $\frac{Y_1}{Y_1 + Y_2}$ obtenemos $\frac{S \cdot C \cdot a + \frac{b}{S\cdot L} + \frac{d}{R}}{S \cdot C \cdot a + \frac{b}{S\cdot L} + \frac{d}{R}+S \cdot C \cdot (1-a) + \frac{1-d}{R} + \frac{1-b}{S \cdot L}}$. Operando algebraicamente podemos reducir esta expresión a $\frac{S \cdot C \cdot a + \frac{b}{S\cdot L} + \frac{d}{R}}{S \cdot C + \frac{1}{S \cdot L} + \frac{1}{R}}$. Multiplicamos al numerador y denominador por $\frac{S}{C}$ y tenemos $T(S) = \frac{S^2 \cdot a+ \frac{b}{L\cdot C} + S\cdot \frac{d}{R\cdot C}}{S^2 + \frac{1}{L \cdot C} + \frac{1}{R\cdot C}}$. Finalmente sacando factor común $a$ en el numerador podemos expresar $T(s)$ como $a\cdot \frac{S^2 + S\cdot \frac{d}{a}\cdot \frac{1}{R\cdot C}+ \frac{b}{a}\cdot \frac{1}{L\cdot C}}{S^2 + S\cdot \frac{1}{R\cdot C} + \frac{1}{L\cdot C}}$. Si comparamos el resultado con una función bicuadrática general de la forma $\frac{A_2\cdot S^2 + A_1 \cdot S + A_0}{B_2\cdot S^2 + B_1 \cdot S + B_0}$ normalizada (Con coeficientes principales 1 en numerador y denominador) de la forma $(\frac{A_2}{B_2})\cdot \frac{S^2 + \frac{A_1}{A_2} \cdot S + \frac{A_0}{A_2}}{S^2 + \frac{B_1}{B_2} \cdot S + \frac{B_0}{B_2}}$, forma muy similar a la pedida por el enunciado $T(s)=k⋅\frac{S^2+S⋅\frac{ω_{0Z}}{Q_Z}+ω_{0Z}^2}{S^2+S⋅\frac{ω_{0P}}{QP}+ω_{0P}^2}$

#### 2. Parametrizar los polinomios de segundo orden, hallando los parámetros $k, \omega_{Z0}, \omega_{P0}, Q_Z, Q_P$ en función de los elementos de la red.

Los parametros pedidos se refieren a la Ganancia del circuito ($k$), Frecuencia Natural de los Ceros ($\omega_{Z0}$, frecuencia donde la respuesta del filtro tiende a 0), Frecuencia Natural de los Polos ($\omega_{P0}$, frecuencia donde ocurre la máxima resonancia si el sistema no está amortiguado), Factor de Calidad de los Ceros ($Q_Z$, mide la anchura de la banda de rechazo) y el Factor de Caliad de los Polos ($Q_P$, mide el amortiguamiento del filtro). Para cumplir la consigna conviene comparar las dos ecuaciones bicuadráticas obtenidas y comparar sus coeficientes. Podemos asociar a la constante $k$ con $a$, a $\frac{d}{a.R.C}$ lo asociamos con $\frac{\omega_{0Z}}{Q_Z}$, $\frac{a}{b}\cdot \frac{1}{L.C}$ con $\omega_{0Z}^2$, $\frac{1}{R.C}$ con $\frac{\omega_{0P}}{Q_P}$ y $\frac{1}{L.C}$ con $\omega_{0Z}^2$.

En este punto claramente las frecuencias de corte y de rechazo $\omega_{0P} = \omega_{0Z} = 1$, mientras que los factores de calidad $Q_Z$ y $Q_P$ valen $5$ y $\frac{\sqrt{2}}{2}$ respectivamente.

#### 4. Dibujar el diagrama de polos y ceros de H(s). Indicar la localización de polos y ceros en función de los parámetros. 

Realizar el diagrama de polos y ceros implica encontrar los ceros del numerador y denominador de la función de transferencia T(s). Los ceros del numerador serán ceros de la función y los ceros del denominador serán polos de la función. Para hallarlos y ubicarlos en el plano complejo utilizamos la herramienta TransferFunction de la libreria Control.matlab. La función TransferFunction toma como argumentos dos arrays bidimensionales, el primero los coeficientes del numerador y el segundo los coeficientes del denominador. Con esta información TransferFunction devuelve una función de transferencia caracterizada por sus coeficientes en la forma de una matriz de coeficientes. Esta matriz es usada como argumento para otra función de la libreria Control llamada Pzmap, que usa esta información para plotear el diagrama de polos y ceros asociado a la función de Transferencia. A continuación vemos las graficas correspondientes a los diagramas de polos y ceros de las tres funciones de transferencia a, b y c. Las cruces indican polos, los circulos indican ceros.

```python

import numpy as np
import matplotlib.pyplot as plt 
import control.matlab as ctrl
import scipy.signal as sig

#Coeficientes de T(s)

#a)
num_a = [1, 0, 9]
den_a = [1, np.sqrt(2), 1]
T_a = ctrl.TransferFunction(num_a, den_a)

#b)
num_b = [1, 0, 1/9]
den_b = [1, 1/5, 1]
T_b = ctrl.TransferFunction(num_b, den_b)

#c)
num_c = [1, 1/5, 1]
den_c = [1, np.sqrt(2), 1]
T_c = ctrl.TransferFunction(num_c, den_c)
```

```python
#Diagrama de Polos y Ceros para a) 
plt.figure(1)
poles_a, zeros_a = ctrl.pzmap(T_a, plot=True)
plt.title('a) Polos y Ceros de $T_a(s)$')
plt.grid(True)
plt.xlabel('Eje Real')
plt.ylabel('Eje Imaginario')
```

```python
#Diagrama de Polos y Ceros para b) 
plt.figure(2)
poles_b, zeros_b = ctrl.pzmap(T_b, plot=True)
plt.title('b) Polos y Ceros de $T_b(s)$')
plt.grid(True)
plt.xlabel('Eje Real')
plt.ylabel('Eje Imaginario')
```

```python
#Diagrama de Polos y Ceros para c) 
plt.figure(3)
poles_c, zeros_c = ctrl.pzmap(T_c, plot=True)
plt.title('c) Polos y Ceros de $T_c(s)$')
plt.grid(True)
plt.xlabel('Eje Real')
plt.ylabel('Eje Imaginario')
plt.tight_layout() 
plt.show()
```

#### 5. Calcular y graficar la respuesta en módulo y fase, indicando la respuesta asintótica donde corresponda 

Para calcular la respuesta en módulo de una función de transferencia hay que llevarla al dominio de fracuencias desde el dominio de Laplace. Esto es, sustituir la variable compleja S por $j\omega$. Luego se calcula el módulo de la función $T(j\omega)$ como el cociente entre los modulos del numerador y del denominador de $T(s)$. La respuesta en fase, por otro lado, se calcula como la diferencia entre las fases del numerador y del denominador, siendo la fase de un numero complejo $Z = a + jb$ el arcotangente de su parte imaginaria sobre su parte real $arctan\frac{b}{a}$.
En nuestras funciones podemos calcular las respuestas en modulo de $T(s)$ como:

a) $\left |T(j\omega)\right |= \left |\frac{(j\omega)^2 + 9}{(j\omega)^2 + (j\omega).\sqrt{2} + 1}\right |$

$\left |T(j\omega)\right |= \frac{\left |9 - \omega^2 \right |}{\sqrt{1+\omega^4}}$

b) $\left |T(j\omega)\right |= \left |\frac{(j\omega)^2 + \frac{1}{9}}{(j\omega)^2 + (j\omega).\frac{1}{5} + 1}\right |$

$\left |T(j\omega)\right |= \frac{\left |\frac{1}{9} - \omega^2 \right |}{\sqrt{1 -\frac{49\omega^2}{25} +\omega^4}}$

c) $\left |T(j\omega)\right |= \left |\frac{(j\omega)^2+ (j\omega).\frac{1}{5} + 1}{(j\omega)^2 + (j\omega).\sqrt{2}+ 1}\right |$

$\left |T(j\omega)\right |= \frac{\sqrt{1 -\frac{49\omega^2}{25} +\omega^4}}{\sqrt{1+\omega^4}}$

y las respuestas en fase como $\angle T(j\omega) = \angle (\text{Numerador}) - \angle (\text{Denominador})$:

a) El numerador de $T_a(j\omega)$ es un numero real, por lo tanto su respuesta en fase será 0 si el numerador es positivo y $\pm \pi$ si es negativo. La respuesta en fase es nula en la frecuencia $\omega_{0Z} = 3$. Entonces:

$\angle (\text{Num}_a) = \begin{cases} 0^\circ & \text{si } \omega < 3 \\ \pm \pi^\circ & \text{si } \omega > 3 \end{cases}$

$\angle (\text{Den}_a) = \arctan\left(\frac{\omega\sqrt{2}}{1 - \omega^2}\right)$

$\angle T_a(j\omega) = \begin{cases} 0^\circ - \arctan\left(\frac{\omega\sqrt{2}}{1 - \omega^2}\right) & \text{si } \omega < 3 \\ \pm 180^\circ - \arctan\left(\frac{\omega\sqrt{2}}{1 - \omega^2}\right) & \text{si } \omega > 3 \end{cases}$

b) Nuevamente el numerador es un numero real puro, por lo que su fase dependerá del signo. 

$\angle (\text{Num}_b) = \begin{cases} 0^\circ & \text{si } \omega < 1/3 \\ \pm 180^\circ & \text{si } \omega > 1/3 \end{cases}$

$\angle (\text{Den}_b) =\arctan\left(\frac{\omega/5}{1 - \omega^2}\right)$

$\angle T_b(j\omega) = \begin{cases} 0^\circ - \arctan\left(\frac{\omega/5}{1 - \omega^2}\right) & \text{si } \omega < 1/3 \\ \pm 180^\circ - \arctan\left(\frac{\omega/5}{1 - \omega^2}\right) & \text{si } \omega > 1/3 \end{cases}$

c) 

$\angle (\text{Num}_c) = \arctan\left(\frac{\omega/5}{1 - \omega^2}\right)$

$\angle (\text{Den}_c) =  \arctan\left(\frac{\omega\sqrt{2}}{1 - \omega^2}\right)$

$\angle T_c(j\omega) = \arctan\left(\frac{\omega/5}{1 - \omega^2}\right) - \arctan\left(\frac{\omega\sqrt{2}}{1 - \omega^2}\right)$

```python
omega = np.logspace(-1, 1, 100)


w_outa, H_jwa = sig.freqresp((num_a, den_a), w=omega)      
magnitud_lineal_a = np.abs(H_jwa)
fase_rads_a = np.angle(H_jwa)

w_outb, H_jwb = sig.freqresp((num_b, den_b), w=omega)      
magnitud_lineal_b = np.abs(H_jwb)
fase_rads_b = np.angle(H_jwb)

w_outc, H_jwc = sig.freqresp((num_c, den_c), w=omega)      
magnitud_lineal_c = np.abs(H_jwc)
fase_rads_c = np.angle(H_jwc)


plt.figure(4)
plt.subplot(1,2,1)
plt.plot(omega,fase_rads_a/np.pi)
plt.title('Respuesta en fase de $T_a(s)$')
plt.xlabel('$\omega$(rads)')
plt.ylabel('Fase de $T(j\omega)$ en $\pi$ radianes')
plt.grid()

plt.subplot(1,2,2)
plt.plot(omega, magnitud_lineal_a)
plt.title('Respuesta en modulo de $T_a(\omega)$ ')
plt.xlabel('$\omega$(rads)')
plt.ylabel('$| T_a(j\omega )|$')
plt.grid()



plt.figure(5)
plt.subplot(1,2,1)
plt.plot(omega,fase_rads_b/np.pi)
plt.title('Respuesta en fase de $T_b(\omega)$')
plt.xlabel('$\omega$(rads)')
plt.ylabel('Fase de $T_b(j\omega)$ en $\pi$ radianes')
plt.grid()

plt.subplot(1,2,2)
plt.plot(omega, magnitud_lineal_b)
plt.title('Respuesta en modulo de $T_b(\omega)$ ')
plt.xlabel('$\omega$(rads)')
plt.ylabel('$| T_b(j\omega )|$')
plt.grid()


plt.figure(6)
plt.subplot(1,2,1)
plt.plot(omega,fase_rads_c/np.pi)
plt.title('Respuesta en fase de $T_c(j\omega)$')
plt.xlabel('$\omega$(rads)')
plt.ylabel('Fase de $T_c(j\omega)$ en $\pi$ radianes ')
plt.grid()

plt.subplot(1,2,2)
plt.plot(omega, magnitud_lineal_c)
plt.title('Respuesta en modulo de $T_c(s)$ ')
plt.xlabel('$\omega$(rads)')
plt.ylabel('$| T_c(j\omega )|$')
plt.grid()



plt.show()

```

Podemos notar observando los distintos graficos que a frecuencias altas los filtros planteados por las T(s) tienen una respuesta asintotica que tiende a 1 en su respuesta en modulo y a 0 en fase.
